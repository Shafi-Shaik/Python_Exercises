```python
import pandas as pd
```

```python
df1=pd.DataFrame({'users':['shafi','ravi','sumit','shafi'],
                'day1':['p','p','p','f'],
                 'day2':['f','p','f','p'],
                 'day3':['p','p','f','p']
                 }    )

finaldf=pd.DataFrame({'users':['ravi','shafi','sumit'],
                'p':[3,4,1],
                 'f':[0,2,2]
                 
                 }    )
print( ' \t\tTransform this :\n\n ', df1 ,'\n'*2 ,'\t\tlike  this:\n \n',finaldf)
```

```python
df1
```

```python
df1=df1.melt(id_vars=["users"], 
        var_name="days", 
        value_name="value")
df1
```

```python
df1=df1.groupby(['users','value']).count().reset_index()
df1
```

## using pivot_table(): which can handle duplicates and has aggregate functions 

```python
df2=df1.pivot_table(index=['users'],columns=['value'],values=['days'],aggfunc=['sum'],fill_value=0).reset_index(level=0)
df2
```

## using pivot(): which cannot handle duplicates and does not have aggregate functions

```python
df2=df1.pivot(index=['users'],columns=['value'],values=['days']).fillna(0).reset_index()
df2
```
